```python
def hexbin(endereco):

    b = ""

    for i in range(len(endereco)):

        if endereco[i] == "2":
            b += "0010"
        elif endereco[i] == "3":
            b += "0011"
        elif endereco[i] == "6":
            b += "0110"
        elif endereco[i] == "9":
            b += "1001"
        elif endereco[i] == "B":
            b += "1011"
        elif endereco[i] == "C":
            b += "1100"
        elif endereco[i] == "F":
            b += "1111"

    return b


endereco = hexbin("3FC92B6")
print("Endereço completo: ", endereco)

bitsE = 0
for i in range(len(endereco)):
    bitsE += 1


import math

KB = 2 ** 10
MB = 2 ** 20

larguraCacheByte = 32
cCache = 128 * KB

qtdLinha = cCache / larguraCacheByte
cMP = 2 ** bitsE

qtdBloco = cMP / larguraCacheByte
qtdBlocoLinha = qtdBloco / qtdLinha

larguraCampoTag = math.log2(qtdBlocoLinha)

bitsLinha = math.log2(qtdLinha)
bitsByte = math.log2(larguraCacheByte)

formatoEndereco = larguraCampoTag + bitsLinha + bitsByte

Ltag = ""
Llinha = ""
Lbyte = ""
for i in range(len(endereco)):
    if i >= 0 and i < larguraCampoTag:
        Ltag = Ltag + endereco[i]
    elif i >= larguraCampoTag and i < (larguraCampoTag + bitsLinha):
        Llinha = Llinha + endereco[i]
    elif i >= (larguraCampoTag + bitsLinha):
        Lbyte = Lbyte + endereco[i]

print(
    "\nFormato de endereço da cache:\n| Tag   |",
    Ltag,
    "\n| Linha |",
    Llinha,
    "\n| Byte  |",
    Lbyte,
    "\nLinha para acesso: ",
    Llinha,
)


print(
    "\nFormato de endereço da cache:\n| Tag   |",
    larguraCampoTag,
    "\n| Linha |",
    bitsLinha,
    "\n| Byte  |",
    bitsByte,
)

print("")
print("Informações da memória: ")
print("")

print("A capacidade da MP é de: ", cMP / MB, "MB")
print("A quantidade de linha é de: ", qtdLinha / KB, "K")
print("A quantidade de bloco é: ", qtdBloco / MB, "M")
print("A quantidade de bloco por linha é de:", qtdBlocoLinha / KB, "K")
print("A largura do campo Tag é de: ", larguraCampoTag, " bits")
print("A quantidade de bits da linha é de: ", bitsLinha, " bits")
print("A quantidade de bits do Byte: ", bitsByte, " bits")
print("O formato do endereço é de: ", formatoEndereco, " bits")
```
