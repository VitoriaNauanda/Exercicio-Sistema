```python
def hexbin(endereco):
    
    b = ""
    
    for i in range(len(endereco)):
     
        if endereco[i] == "2": 
              b += "0010"
        elif endereco[i] == "3":
        	 b += "0011"
        elif endereco[i] == "6": 
             b += "0110"
        elif endereco[i] == "9":
        	 b += "1001"
        elif endereco[i]== "B":
        	 b += "1011"
        elif endereco[i]== "C":
        	 b += "1100"
        elif endereco[i] == "F": 
             b += "1111"  
        
    return ((b))
endereco = (hexbin("3FC92B6"))
print("Endereço completo: ", endereco)

bitsE = 0
for i in range(len(endereco)):
	bitsE += 1


import math

KB = 2**10
MB = 2**20

cMP = 2**bitsE
cCache = 64 * KB
larguraCacheByte = 32
linhaConjunto = 4

qtdLinha = cCache/ larguraCacheByte
qtdConjunto = qtdLinha / linhaConjunto

t = math.log2(qtdConjunto)
qtdBloco = cMP / larguraCacheByte
qtdBlocoConjunto = qtdBloco /(2**t)

larguraCampoTag = math.log2(qtdBlocoConjunto)
larguraCampoConjunto = math.log2(qtdConjunto)

bitsByte = math.log2(larguraCacheByte)

formatoEndereco = larguraCampoTag + larguraCampoConjunto + bitsByte

Ltag = ""
Lcampo = ""
Lbyte = ""
for i in range(len(endereco)):
    if(i>=0 and i<larguraCampoTag):
        Ltag = Ltag + endereco[i]
    elif(i>=larguraCampoTag and i<(larguraCampoTag+ larguraCampoConjunto)):
        Lcampo = Lcampo + endereco[i]
    elif(i>=(larguraCampoTag+ larguraCampoConjunto)):
         Lbyte = Lbyte + endereco[i]

print("\nFormato de endereço da cache:\n| Tag   |",Ltag,"\n| Campo |",Lcampo,
      "\n| Byte  |",Lbyte,"\nCampo para acesso: ", Lcampo)
              

print("\nFormato de endereço da cache:\n| Tag   |", larguraCampoTag,"\n| Campo |",larguraCampoConjunto,
      "\n| Byte  |",bitsByte)


print(" ")
print("Informações da Memória: ")
print(" ")
print("A capacidade da MP é de: ", cMP /MB, " MB")
print("A quantidade de linha é de: ", qtdLinha / KB, " K")
print("A quantidade de conjunto é de: ", qtdConjunto )
print("A quantidade de bloco é de: ", qtdBloco / MB, " M")
print("A quantidade de bloco por conjunto é de: ", qtdBlocoConjunto/ KB, " K")
print("A largura do campo Tag é de: ", larguraCampoTag, " bits")
print("A largura do campo conjunto é de: ", larguraCampoConjunto, " bits")
print("Campo byte: ", bitsByte, " bits")
print("O formato do endereço é de: ", formatoEndereco, " bits")
```
