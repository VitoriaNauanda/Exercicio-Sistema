```python
def hexbin(endereco):
    
    b = ""
    
    for i in range(len(endereco)):
        if endereco[i] == "2": 
             b += "0010"
        elif endereco[i] == "3":
        	 b += "0011"
        elif endereco[i] == "6": 
             b += "0110"
        elif endereco[i] == "9":
        	 b += "1001"
        elif endereco[i]== "B":
        	 b += "1011"
        elif endereco[i]== "C":
        	 b += "1100"
        elif endereco[i] == "F": 
             b += "1111"
        
    return ((b))
endereco = (hexbin("3FC92B6"))
print("Endereço completo: ", endereco)

bitsE = 0

for i in range(len(endereco)):
	bitsE+= 1
    
print("Tamanho em bits do Endereço: ",  bitsE )

import math

KB = 2**10
MB = 2**20

larguraCacheByte = 32

cCache = 64 * KB
cMP = 2**bitsE 

qtdLinha = cCache / larguraCacheByte
qtdBloco = cMP / larguraCacheByte
qtdBlocoLinha = qtdBloco / qtdLinha

larguraCampoBloco = math.log2(qtdBloco)

bitsByte = math.log2(larguraCacheByte)

formatoEndereco = larguraCampoBloco + bitsByte 

Lcampo = ""
Lbyte = ""
for i in range(len(endereco) ):
    if(i>=0 and i<larguraCampoBloco):
        Lcampo = Lcampo + endereco[i]
    elif(i>=larguraCampoBloco and i<(larguraCampoBloco + bitsByte)):
        Lbyte= Lbyte+ endereco[i]
    elif(i>=(larguraCampoBloco+ bitsByte)):
         Lbyte = Lbyte + endereco[i]

print("\nFormato de endereço da cache:\n| Campo |",Lcampo,
      "\n| Byte  |",Lbyte,"\nCampo para acesso: ", Lcampo)
              

print("\nFormato de endereço da cache:\n| Campo|",larguraCampoBloco,
      "\n| Byte  |",bitsByte)
      
print("")
      
print("Informações da Memória: ")

print(" ")


print("A capacidade da MP é de: ", cMP / MB, " MB")
print("A quantidade de linha é de: ", qtdLinha / KB , " K")
print("A quantidade de bloco é de: ", qtdBloco / MB, " M")
print("A quantidade de bloco por linha é de: ", qtdBlocoLinha / KB, " K")
print("A largura do campo bloco é de: ", larguraCampoBloco, " bits")
print("O campo byte é de: ", bitsByte, " bits")
print("O formato do endereço é de: ", formatoEndereco, " bits")
```
